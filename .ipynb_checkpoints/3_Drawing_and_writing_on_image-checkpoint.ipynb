```python
import cv2
import numpy as np
```

```python
img = cv2.imread("images/joker.jpg", cv2.IMREAD_COLOR)
```

```python
cv2.line(img, (10,10), (50, 50), (255,255,255), 5)
```

```python
cv2.line(img, (50,10), (10,50), (255,255,255), 5)
```

```python
cv2.rectangle(img, (10,10), (50,50), (0,255,255), 5)
```

```python
cv2.circle(img, (400,400), 100, (255,0,0), -1)
```

```python
addH = 100
addW = 100
pts = np.array([[10,400],[300,100],[400,50],[5,2]], np.int32)
pts[:,0] += addH
pts[:,1] += addW
pts = pts.reshape((-1,1,2))
cv2.polylines(img, [pts], True, (0,255,255), 3)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, "Here text", (100, 800), font, 10, (100,255,255), 20, cv2.LINE_AA)
```

```python
cv2.imshow("image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
